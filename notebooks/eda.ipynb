```python
import pyarrow.parquet as pq

def download_file(url: str, local_path: str, chunk_size: int = 1024):
    """Baixa o arquivo da URL de forma segura, em chunks."""
    response = requests.get(url, stream=True)
    response.raise_for_status()  # Levanta erro se status != 200
    with open(local_path, 'wb') as f:
        for chunk in response.iter_content(chunk_size=chunk_size):
            f.write(chunk)

def load_df_from_url(dir_path: str, url: str) -> pd.DataFrame | None:
    dir_path = Path(dir_path)
    dir_path.mkdir(parents=True, exist_ok=True)

    file_name = url.split('/')[-1]
    local_path = dir_path / file_name

    # Baixar apenas se não existir
    if not local_path.exists():
        try:
            download_file(url, local_path)
        except Exception as e:
            print(f"Erro ao baixar o arquivo de {url}: {e}")
            return None

    return load_df(local_path)

# Carregar os DataFrames
df_cadastral = load_df_from_url("./data", f"{github_url}/base_cadastral.parquet")
df_emprestimos = load_df_from_url("./data", f"{github_url}/historico_emprestimos.parquet")
df_submissao = load_df_from_url("./data", f"{github_url}/base_submissao.parquet")
df_parcelas = load_df_from_url("./data", f"{github_url}/historico_parcelas.parquet")
dicionario = load_df_from_url("./data", f"{github_url}/dicionario_dados.csv")
```

```python
def display_missings(df):
  missing_values = df.isna().sum()

  # Filtra apenas as colunas com nulos
  missing_values_filtered = missing_values[missing_values > 0]

  # Calcula a porcentagem
  total_rows = len(df)
  percentage_missing = ((missing_values_filtered / total_rows) * 100).round(1)

  # Cria um novo DataFrame para exibir as informações
  missing_info = pd.DataFrame({
      'Contagem de Nulos': missing_values_filtered,
      'Porcentagem de Nulos (%)': percentage_missing
  })

  missing_info_sorted = missing_info.sort_values(by='Porcentagem de Nulos (%)', ascending=False)

  display(missing_info_sorted)
```

```python
def df_overview(df, name):
  '''
  Visão geral da base de dados
  '''
  print('='*12 + f" Base {name} " + '='*12)
  print("\nDimensão da base:", df.shape)
  print("Amostras duplicadas:", df.duplicated().sum())

  print("\n" + "-"*40)
  print("\nInfo da base:")
  df.info()

  print("\n" + "-"*40)
  print("\nValores Missing:")
  display_missings(df)

  print("\n" + "-"*40)
  print("\n Estatísticas descritivas:")
  # Excluir colunas de ID antes de calcular estatísticas descritivas
  cols_to_exclude = ['id_contrato', 'id_cliente']
  df_for_describe = df.drop(columns=[col for col in cols_to_exclude if col in df.columns], errors='ignore')
  display(df_for_describe.describe().transpose())

  print("\n" + "-"*40)
  print("\n Primeiras linhas:")
  display(df.head(3))
```

## Limpeza das Bases

Descobri que há registros com mesmo id_contrato e mesmo numero_parcela mas com versao_parcela distintos. 

Preparação da base de parcelas:
- selecionar apenas as parcelas com a versão mais recente
- Agregar as parcelas restantes que têm mesmo id_contrato e numero_parcela:
    - somar valor_pago_parcela
    - atribuir o valor mais recente a data_real_pagamento

```python
contagem_versoes = df_parcelas.groupby(['id_contrato', 'numero_parcela']).agg(
    num_versoes_unicas=('versao_parcela', 'nunique')
).reset_index()

# filtrar apenas os casos com mais de uma versão
contagem_versoes = contagem_versoes[contagem_versoes['num_versoes_unicas'] > 1].sort_values(by='num_versoes_unicas', ascending=False)
contagem_versoes.head()
```

```python
reparceladas_mask = df_parcelas.duplicated(subset=['id_contrato', 'numero_parcela'], keep=False)
df_reparceladas = df_parcelas[reparceladas_mask].sort_values(['id_contrato', 'numero_parcela'])

df_reparceladas.head()
```

```python
total_linhas_reparceladas = len(df_reparceladas)

num_parcelas_unicas_reparceladas = df_reparceladas.groupby(['id_contrato', 'numero_parcela']).ngroups

num_contratos_com_reparcelamento = df_reparceladas['id_contrato'].nunique()

print(f"Número total de linhas envolvidas em reparcelamentos: {total_linhas_reparceladas:,}")
print(f"Número de parcelas únicas que foram reparceladas: {num_parcelas_unicas_reparceladas:,}")
print(f"Número de contratos únicos com reparcelamento: {num_contratos_com_reparcelamento:,}")
```

```python
df_parcelas['ultima_versao'] = df_parcelas.groupby(['id_contrato', 'numero_parcela'])['versao_parcela'].transform('max')

df_latest_versions = df_parcelas[df_parcelas['versao_parcela'] == df_parcelas['ultima_versao']].copy()
```

O resuldado abaixo mostra que para uma mesma versão de uma parcela não é possível ter valores previstos diferentes

```python
# Verifica se valor_previsto_parcela é sempre igual para cada grupo de id_contrato e numero_parcela
variacao_valor_previsto = (
    df_latest_versions
    .groupby(['id_contrato', 'numero_parcela'])['valor_previsto_parcela']
    .nunique()
)

num_grupos_com_variacao = (variacao_valor_previsto > 1).sum()
print(f"Número de grupos com valor_previsto_parcela diferente: {num_grupos_com_variacao}")

# Visualizar os casos onde há variação de valor_previsto_parcela
grupos_com_variacao = variacao_valor_previsto[variacao_valor_previsto > 1].index

casos_variacao = df_reparceladas.set_index(['id_contrato', 'numero_parcela']).loc[grupos_com_variacao].reset_index()
display(casos_variacao.head(10))
```

```python
agg_rules = {
    'id_cliente': 'first', # O id_cliente será o mesmo
    'versao_parcela': 'first', # A versão será a mesma
    'data_prevista_pagamento': 'min', # Pegamos a data de vencimento mais antiga da renegociação
    'data_real_pagamento': 'max', # Pegamos a última data de pagamento real
    'valor_previsto_parcela': 'first', # O valor é o mesmo
    'valor_pago_parcela': 'sum' # Somamos os valores pagos
}

df_parcelas_consolidadas = df_latest_versions.groupby(['id_contrato', 'numero_parcela']).agg(agg_rules).reset_index()
```

```python
print(f"Número de parcelas consolidadas: {len(df_parcelas_consolidadas)}")
```
